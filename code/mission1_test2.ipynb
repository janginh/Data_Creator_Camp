```python
from ultralytics import YOLO

# 1) 모델 로드 (yolov8l)
model = YOLO("yolov8l.pt")

# 2) 학습
results = model.train(
    data="data.yaml",  # 데이터셋 yaml 경로
    conf=0.25,      #confidence 0.25
    iou=0.5,        #IoU 0.5
    max_det=100,    #max detections 100
    epochs=100,
    imgsz=512,          
    batch=16,
    workers=4,          # 로더 스레드 수 (환경에 맞게)
    project="runs",     # 결과 저장 루트
    name="yolov8l_chimney",  # 실험명
    exist_ok=False,      # 같은 이름 덮어쓰기 허용
    verbose=True,
    patience=20,        # 20epoch 이후에도 성능 향상이 없을시 종료
    seed=42,
)


```

```python
# 3) 검증(선택: best.pt로 val set 평가)
metrics = model.val(data="data.yaml")

```

```python
# 4) 예측(시각화 저장)
pred = model.predict(
    source="images/valid",  # 이미지/폴더/동영상 가능
    conf=0.25,
    imgsz=512,
    save=True,
    project="runs",
    name="yolov8l_chimney_pred",
    exist_ok=True
)
```
